```python

import matplotlib.pyplot as plt
from scipy.stats import norm, gaussian_kde

# Sample data
data = np.array([1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5, 6, 6, 6, 7, 7, 8])

# Probability Density Function (PDF)
plt.figure(figsize=(8, 5))
kde = gaussian_kde(data)
x_vals = np.linspace(min(data), max(data), 1000)
plt.plot(x_vals, kde(x_vals), label='PDF', color='blue')
plt.title('Probability Density Function (PDF)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

# Probability Mass Function (PMF) - Discrete Approximation
plt.figure(figsize=(8, 5))
unique_values, counts = np.unique(data, return_counts=True)
pmf = counts / sum(counts)
plt.bar(unique_values, pmf, label='PMF', color='green')
plt.title('Probability Mass Function (PMF)')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.show()

# Cumulative Distribution Function (CDF)
plt.figure(figsize=(8, 5))
cdf_vals = np.sort(data)
cdf = np.arange(1, len(cdf_vals) + 1) / len(cdf_vals)
plt.plot(cdf_vals, cdf, label='CDF', marker='o', linestyle='-', color='red')
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()
```

# Implementing Bernouli Distribution

```python
from scipy.stats import bernoulli
p = 0.6
data = bernoulli.rvs(p, size=1000)
counts = [sum(data == 0), sum(data == 1)]
plt.bar([0, 1], counts, color=['red', 'green'], tick_label=['Failure (0)', 'Success (1)'])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution (p=0.6)")
plt.show()
```

# Implimenting Binomial Distribution 

```python
from scipy.stats import binom
n = 10  
p = 0.3 

x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)
plt.bar(x, pmf, color='green', alpha=0.7,edgecolor='black',linewidth=2.0)
plt.xlabel("Number of Successes (Clicks)")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.show()
```

# implimenting Poission Distribution

```python
from scipy.stats import poisson

lambda_ = 5  
x = np.arange(0, 10)  
pmf = poisson.pmf(x, lambda_)

plt.bar(x, pmf, color='white', alpha=0.7, edgecolor='black', linewidth=2.5)  
plt.xlabel("Number of Calls per Minute")
plt.ylabel("Probability")
plt.title(f"Poisson Distribution (λ={lambda_})")
plt.show()
```

# Implimenting NOrmal Distribution

```python
from scipy.stats import norm
mu = 70  
sigma = 10  
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf = norm.pdf(x, mu, sigma)  
plt.plot(x, pdf, color='red', label=f"Normal Distribution (μ={mu}, σ={sigma})")
plt.xlabel("Exam Scores")
plt.ylabel("Probability Density")
plt.title(f"Normal Distribution (μ={mu}, σ={sigma})")
plt.legend()
plt.grid()
plt.show()
```

# Implimenting Standard Normal Distribution (Z-Distribution)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

z = np.linspace(-4, 4, 1000)  
pdf = norm.pdf(z, 0, 1)  

plt.plot(z, pdf, color='blue', label="Standard Normal Distribution (μ=0, σ=1)")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.title("Standard Normal Distribution (Z-Distribution)")
plt.axvline(x=0, color='red', linestyle='--', label="Mean (Z=0)")
plt.legend()
plt.grid()
plt.show()
```

```python

np.random.seed(42)
income_data = np.random.exponential(scale=50, size=1000)

# Plot the original skewed distribution
plt.figure(figsize=(8, 4))
sns.histplot(income_data, bins=50, kde=True, color="brown")
plt.title("Original Income Distribution (Skewed)")
plt.xlabel("Income ($1000)")
plt.ylabel("Frequency")
plt.show()
```

```python

# Perform sampling and calculate means
sample_means = []
n_samples = 1000
sample_size = 30  

for _ in range(n_samples):
    sample = np.random.choice(income_data, sample_size)
    sample_means.append(np.mean(sample))

# Plot the sampling distribution of the sample mean
plt.figure(figsize=(8, 4))
sns.histplot(sample_means, bins=50, kde=True, color="green")
plt.title(f"Sampling Distribution of Sample Mean (n={sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
